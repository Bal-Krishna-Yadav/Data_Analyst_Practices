```python
df['Calories (kcal)'].hist(bins=20, color='skyblue', edgecolor='black')
plt.xlabel('Calories')
plt.ylabel('Frequency')
plt.title('Distribution of Calories')
plt.show()
```

```python


top10 = df.sort_values('Calories (kcal)', ascending=False).head(10)

# Create the plot without setting the palette
plt.figure(figsize=(12, 6))
ax = sns.barplot(x='Food Name', y='Calories (kcal)', data=top10)

# Custom colors for each bar
colors = sns.color_palette("Set2", len(top10))

# Apply colors to each bar individually
for bar, color in zip(ax.patches, colors):
    bar.set_color(color)

plt.xlabel('Food Name')
plt.ylabel('Calories')
plt.title('Top 10 High-Calorie Foods')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
sns.scatterplot(x='Protein (g)', y='Calories (kcal)', data=df,color="red")
plt.xlabel('Protein(g)')
plt.ylabel('Calories(kcl)')
plt.title('Protein(g) vs Calories(kcl)')
plt.show()
```

```python
select_cols = ['Calories (kcal)', 'Protein (g)','Total Fat (g)','Carbohydrates (g)','Sugar (g)','Fiber (g)','Sodium (mg)','Cholesterol (mg)',
               'Calcium (mg)','Iron (mg)','Potassium (mg)','Vitamin C (mg)']
numeric_df = df[select_cols]
```

```python
numeric_df.head(5)
```

```python
plt.figure(figsize=(10, 6))
sns.heatmap(numeric_df.corr(), annot=True,cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Matrix')
plt.show()
```

```python

# Step 1: Sort the dataset by Iron (mg) descending and take top 10
top5 = df.sort_values('Iron (mg)', ascending=False).head(5)

# Step 2: Set Food Name as index and get Iron values
iron_values = top5.set_index('Food Name')['Vitamin C (mg)']

# Step 3: Plot the pie chart
iron_values.plot.pie(
    autopct='%1.1f%%',
    figsize=(8, 8),
    fontsize=12,
    startangle=140
)

plt.title('Top 10 Iron-Rich Foods Distribution')
plt.ylabel('')  # Removes default y-axis label
plt.show()
```

```python
# CSV read
df = pd.read_csv(r"C:\Users\Bal Krishna Yadav\Downloads\Food_Nutrient_Dataset_200.csv")

# Strip whitespace from column names
df.columns = df.columns.str.strip()

# Print all columns to verify
print("📌 Columns in your CSV:", df.columns.tolist())

def search_food(food_name):
    if 'Food Name' in df.columns:
        result = df[df['Food Name'].str.lower() == food_name.lower()]
        
        if not result.empty:
            print("\n✅ Food Details Found:\n")
            for column in result.columns:
                print(f"{column}: {result.iloc[0][column]}")
        else:
            print("\n❌ Sorry! Food not found in the dataset.")
    else:
        print("⚠️ 'Food Name' column not found. Please check your CSV.")

# User input
food = input("🔍 Enter food name to search: ")
search_food(food)
```
